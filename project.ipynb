- Porcentaje personas sobrevivieron vs murieron

```python
muertos = df.loc[df["Survived"] == 0].shape[0]
vivos = df.loc[df["Survived"] == 1].shape[0]
total = muertos + vivos
print("Murieron: " + str(muertos))
print("Sobrevivieron: " + str(vivos))
print ("Total de personas: " + str(total))
#He puesto que solo salgan 2 decimales si no quedaba muy largo
print(f"Porcentaje muertos: {(muertos / total) * 100:.2f}% Porcentaje vivos: {(vivos / total) * 100:.2f}%")
```
